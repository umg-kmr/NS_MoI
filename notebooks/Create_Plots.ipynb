```python
import pandas as pd
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
```

```python
#Load the output file
eos_name = 'ALF2'
cols = ['M','R','L','I'] #L is \Lambda
df = pd.read_csv('../output/'+eos_name+'/MoIL.dat',names=cols)
```

```python
#Mass vs Radius Plot

plt.plot(df.R,df.M)
plt.xlabel('Radius(Km)')
plt.ylabel('$M/M_\odot$')
plt.grid()
plt.show()
```

```python
#\Lambda vs M plot

plt.plot(df.M,df.L)
plt.xlabel('$M/M_\odot$')
plt.ylabel('$\Lambda$')
# Limits for the axis
# plt.ylim(0,3500)
# plt.xlim(0.9)
plt.grid()
plt.show()
```

```python
#I vs M plot

plt.plot(df.M,df.I)
plt.xlabel('$M/M_\odot$')
plt.ylabel('I (kg $m^2$)')
plt.grid()
plt.show()
```

```python
#I_45/M plot with \Lambda constraints from GW170817

plt.plot(df.M,df.I.mul(1/(1e38*df.M)))
plt.xlabel('$M/M_\odot$')
plt.ylabel('$I_{45}/M \ (g \ cm^2 \ M_{\odot}^{-1})$')
#Plotting constraints on \Lambda from GW170817
for i in range(len(df.M)):
    #50% Probability
    if df.L[i]<=1000:
        plt.plot(df.M[i],df.I.mul(1/(1e38*df.M))[i],'co',markersize=4,alpha=.25,markevery=5)
    #90% Probability
    elif df.L[i]>1000 and df.L[i]<=2700:
        plt.plot(df.M[i],df.I.mul(1/(1e38*df.M))[i],'bo',markersize=4,alpha=.25)
plt.grid()
plt.show()
```
